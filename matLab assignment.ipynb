Matplotlib is a graphics library used for generating 2D and 3D figures. It also provides an API designed to be similar to MATLAB. Here, I will use matplotlib along with NumPy and other libraries to create a series of graphs and charts for subsets of data.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

First, we will use Pandas to import a .csv file, then we will create seperate Numpy arrays for the ages of the male marathon runners and the female marathon runners.

```python
marathon = pd.read_csv("marathon-data.csv")
maleAges = marathon[marathon['gender'] == 'M']['age']
femaleAges = marathon[marathon['gender'] == 'W']['age']
```

First, we import the .csv with the function `read_csv` which we imported as part of the Pandas library. Then we sliced the array. We identify the column we wish to slice by - gender - and what we are looking for - man or woman, identified in the file by the character 'M' or 'W". Then we identify the column we wish to be included in the array - age. With our two arrays, we will then use matplotlib to create a pair of histograms.

We begin by making a function of our own, where we will use the `hist` function to create a histogram with a title, then we will call it twice.

```python
def marathonHist(ages, name):
    bins = np.linspace(0,80,50)
    plt.hist(ages,alpha = 0.5, density = True, label = name)
    plt.legend(loc = 'upper right')
    plt.show()
    
marathonHist(maleAges, "Male Ages")
marathonHist(femaleAges, "Female Ages")
```

We will also use matplot lib to create a scatterplot with an inset graph that zooms in on a particular section of the data. As before, we will use pandas to import a .csv and then convert the DataFrame to a numpy array before using matplotlib to create a graph. 

We will make two sets of axes to create a main graph and an insert. (See file SCIPY ICU Beds for a more detailed explanation of the initial graph creation). This is different from previous attempts because of the insert - first, we must create the figure using thr `figure` function, then add two separate sets of axes using the `add_axes` function. The listed numbers are the position of the axes on a scale from 0 to 1.

```python
data = pd.read_csv("kaggleBeds.csv")
popData = data['Total Population'].to_numpy()
bedData = data["ICU Beds"].to_numpy()


fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes2 = fig.add_axes([0.4, 0.55, 0.4, 0.3])


axes1.scatter(popData,bedData,c='b',marker='*')
axes1.set_xlabel('Number of ICU Beds', fontsize=16)
axes1.set_ylabel('Total Population', fontsize=16)
axes1.set_title('Number of ICU Beds vs. Total Population',fontsize=20)

#axes2.set_xlim(0,0.5)
axes2.set_ylim(0,1000)
axes2.scatter(popData,bedData, c ='g', marker = '*')
axes2.set_xlabel('Number of ICU Beds', fontsize=10)
axes2.set_ylabel('Total Population', fontsize=10)
```
